```python
import pandas as pd
```

```python
file1=pd.read_csv("file1.csv")
file2=pd.read_csv("file2.csv")
file3=pd.read_csv("file3.csv")
```

```python
file1=file1.rename(columns={"Customer":"ID","ST":"State","GENDER":"Gender","Customer Lifetime Value":"CLV","Monthly Premium Auto":"MPA","Number of Open Complaints":"NoC","Policy Type":"PT","Vehicle Class":"VC","Total Claim Amount":"TCA"})
file2=file2.rename(columns={"Customer":"ID","ST":"State","GENDER":"Gender","Customer Lifetime Value":"CLV","Monthly Premium Auto":"MPA","Number of Open Complaints":"NoC","Total Claim Amount":"TCA","Policy Type":"PT","Vehicle Class":"VC"})
file3=file3.rename(columns={"Customer":"ID","Customer Lifetime Value":"CLV","Monthly Premium Auto":"MPA","Number of Open Complaints":"NoC","Total Claim Amount":"TCA","Policy Type":"PT","Vehicle Class":"VC"})
```

```python
file2=file2[["ID","State","Gender","Education","CLV","Income","MPA","NoC","PT","VC","TCA"]]
file3=file3[["ID","State","Gender","Education","CLV","Income","MPA","NoC","PT","VC","TCA"]]
```

```python
customers=pd.concat([file1,file2,file3],axis=0)
```

```python
customers
```

```python
customers=customers.reset_index(drop=True)
```

```python
customers
```

```python
customers.dtypes
```

```python
customers=customers.drop(["Education","NoC"],axis=1)
```

```python
customers["CLV"]=customers["CLV"].astype(object)
```

```python
#I think instead of multiplying, i'm gonna be dividing them by 100 to get the decimals
customers["CLV"] = list(map(lambda x: x[:-1], customers['CLV'].values))
#customers["CLV"] = [float(x) for x in customers['CLV'].values]
```

```python
def p2f(x):
    return float(x.strip('%'))/100
```

```python
for i in customers["CLV"]:
    customers["CLV"]=p2f(i)
```

```python
customers["CLV"].astype("object").apply(p2f)
```

```python
customers=customers.drop_duplicates()
```

```python
customers["BIGIncome"]= customers["Income"]>0
```
